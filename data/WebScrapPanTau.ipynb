```python
import re

# 1. Menghapus whitespace di awal dan akhir teks
df_fufufafa['Title'] = df_fufufafa['Title'].str.strip()
df_fufufafa['Content'] = df_fufufafa['Content'].str.strip()

# 2. Menghapus karakter non-ASCII dari konten (misalnya simbol atau karakter aneh)
df_fufufafa['Content'] = df_fufufafa['Content'].apply(lambda x: re.sub(r'[^\x00-\x7F]+', ' ', x))

# 3. Menghapus pola teks yang tidak diinginkan, seperti "Sumber gambar" atau "Tanggal"
df_fufufafa['Content'] = df_fufufafa['Content'].apply(lambda x: re.sub(r'(Sumber gambar|Tanggal|Foto:)', '', x, flags=re.IGNORECASE))

# 4. Menghapus spasi berlebih atau baris kosong
df_fufufafa['Content'] = df_fufufafa['Content'].apply(lambda x: re.sub(r'\s+', ' ', x).strip())

# 5. (Opsional) Convert to lower case jika diperlukan normalisasi lebih lanjut
df_fufufafa['Content'] = df_fufufafa['Content'].str.lower()

# 6. Menghapus tanda baca yang tidak diperlukan
df_fufufafa['Content'] = df_fufufafa['Content'].apply(lambda x: re.sub(r'[^\w\s]', '', x))
```

```python
# Menghapus duplikasi paragraf dalam konten
df_fufufafa['Content'] = df_fufufafa['Content'].apply(lambda x: ' '.join(sorted(set(x.split()), key=x.index)))

# Menggabungkan kalimat atau baris yang dipisahkan ke dalam satu paragraf
df_fufufafa['Content'] = df_fufufafa['Content'].apply(lambda x: ' '.join(x.splitlines()))
```

```python
from bs4 import BeautifulSoup

# Fungsi untuk membersihkan tag HTML yang tidak diinginkan
def clean_html(raw_html):
    soup = BeautifulSoup(raw_html, 'html.parser')
    for tag in soup(['script', 'style', 'iframe']):
        tag.extract()  # Hapus tag script, style, dan iframe
    return soup.get_text(separator=' ', strip=True)

# Terapkan ke kolom 'Content'
df_fufufafa['Content'] = df_fufufafa['Content'].apply(clean_html)
```

```python
# Menghapus pola teks yang mengandung "credit" atau "sumber"
def remove_patterns(text):
    patterns = ['credit:', 'sumber:', 'iklan']
    for pattern in patterns:
        text = re.sub(rf'{pattern}.*', '', text, flags=re.IGNORECASE)
    return text

# Terapkan ke 'Content'
df_fufufafa['Content'] = df_fufufafa['Content'].apply(remove_patterns)
```

```python
# Memecah paragraf yang terlalu panjang menjadi lebih kecil berdasarkan jumlah kata
def split_paragraph(paragraph, word_limit=50):
    words = paragraph.split()
    return [' '.join(words[i:i + word_limit]) for i in range(0, len(words), word_limit)]

# Terapkan ke kolom 'Content'
df_fufufafa['Content'] = df_fufufafa['Content'].apply(lambda x: ' \n'.join(split_paragraph(x, 50)))
```
